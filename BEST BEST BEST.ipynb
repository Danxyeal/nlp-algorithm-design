```python
from bs4 import BeautifulSoup
import pprint

pp = pprint.PrettyPrinter(indent=2)
```

```python
def sift_tags_down(line):
    temp = ''
    for char in line:
        if char == '<':
            temp += '\n<'
        else:
            temp += char
    return temp
    
def remove_meaningless_newlines(line):
    for i in range(0, len(line)):
        if not i+2 > len(line):
            if line[i] == '\n' and line[i+1] == '<':
                return line[i]
            elif line[i] == '\n':
                return ''
            else:
                return line[i]

def strip_newlines(Lines):
    line_list = []
    for line in Lines:
        line = remove_meaningless_newlines(line)
        line = sift_tags_down(line)
        line_list.append(line)
    print(line_list)
    return line_list

def get(filename):
    with open(filename, 'r') as html_file:
        return html_file.readlines()
    
def read(line):
    soup = BeautifulSoup(line, 'html.parser')
    return soup
#    return [s.name for s in soup.find_all() if not s == None]

def read_tags(soup):
    return [tag.name for tag in soup.find_all()]
def read_attrs(soup):
    return [tag.attrs for tag in soup.find_all()]
def read_text(soup):
    return [tag.text for tag in soup.find_all()]

def remove_char(rm, line_string):
    line = ''
    for char in line_string:
        if char == rm:
            line += ''
        else:
            line += char
    return line
        
        
html_lines = get('proj_files/asus_mb16ac.html')
run = [remove_char('\t', line) for line in html_lines]
wax = [remove_char('\n', line) for line in run]
#lol = [remove_char(' ', line) for line in wax]
soups = [read(line) for line in wax]
tags = [read_tags(soup) for soup in soups]
attrs = [read_attrs(soup) for soup in soups]
text = [read_text(soup) for soup in soups]


'''for i in range(0, len(wax)):
    print(soups[i])
    print(type(soups[i]))
    print('###############################################################')'''

for i in range(0, len(tags)):
    print(f'HTML_TAG{",".join(tags[i])}, {attrs[i]}, {"".join(text[i])}')
```
