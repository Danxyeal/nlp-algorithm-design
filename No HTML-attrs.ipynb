```python
from bs4 import BeautifulSoup
import pprint

def get(file):
    with open(file, 'r') as html_file:
        lines = html_file.readlines()
        return [l.replace('\n','') for l in lines]
    
def soup_lines(lines):
    for text in lines:
        return (BeautifulSoup(text, 'html.parser') for text in lines)

read_tags = lambda soup: [tag.name for tag in soup.find_all()]
read_attrs = lambda soup: [tag.attrs for tag in soup.find_all() if not tag.attrs == {}]
read_text = lambda soup: [tag.text for tag in soup.find_all()]

def join_parts(tag, attr, text):
    return {','.join(tag): attr, 'raw text': text}
       
def join_tags_to_text(tag, text):
    if not tag and not text:
        return False
    tag = ''.join(tag)
    return f'{tag or None}: {text or None}'

def destructure_noisy_html(html_lines):

    lines = [line.replace('\t','') for line in html_lines]
    tags = [read_tags(soup) for soup in soup_lines(lines)]
    attrs = [read_attrs(soup) for soup in soup_lines(lines)]
    text = [soup.text for soup in soup_lines(lines)]
    #return (list(map(join_parts, tags, attrs, text))) 
    qline = list(map(join_tags_to_text, tags, text))
    return [l for l in qline if l]

if __name__ == ('__main__'):
    noisy_html = get('specs-chunk.html')
    less_noisy = destructure_noisy_html(noisy_html)
    quieter = [line for line in less_noisy if not line == '[]: ']
    #quieter = [line for line in less_noisy if not line == ', [], '] 
    #impler = join_text_lines(quieter)
    for l in quieter:
        print(l)
```
