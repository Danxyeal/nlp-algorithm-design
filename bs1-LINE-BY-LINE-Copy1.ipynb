```python
from bs4 import BeautifulSoup
import pprint

pp = pprint.PrettyPrinter(indent=2)
```

```python
def sift_tags_down(line):
    temp = ''
    for char in line:
        if char == '<':
            temp += '\n<'
        else:
            temp += char
    return temp
    
def remove_meaningless_newlines(line):
    for i in range(0, len(line)):
        if not i+2 > len(line):
            if line[i] == '\n' and line[i+1] == '<':
                return line[i]
            elif line[i] == '\n':
                return ''
            else:
                return line[i]

def strip_newlines(Lines):
    line_list = []
    for line in Lines:
        line = remove_meaningless_newlines(line)
        line = sift_tags_down(line)
        line_list.append(line)
    print(line_list)
    return line_list

def get_list_file(filename):
    with open(filename, 'r') as html_file:
        return html_file.readlines()
```
